```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
sns.set_theme(color_codes=True)
```

```python
#loading data into dataframe
df = pd.read_csv('data/application_train.csv')

df.head()
```

```python
#calculating missing rate for each column and sorting
missing_rate = (df.isnull().sum() * 100) / df.shape[0]
missing_rate.sort_values(ascending=False)
```

```python
#filling na values with mean value
avg_columns = [col for col in df.columns if 'AVG' in col]
for col in avg_columns:
    avg_value = df[col].mean()
    df[col].fillna(avg_value, inplace=True)
```

```python
#filling na values with median value
avg_columns = [col for col in df.columns if 'MEDI' in col]
for col in avg_columns:
    avg_value = df[col].median()
    df[col].fillna(avg_value, inplace=True)
```

```python
#filling na values with mode value
avg_columns = [col for col in df.columns if 'MODE' in col]
for col in avg_columns:
    avg_value = df[col].mode()[0]
    df[col].fillna(avg_value, inplace=True)
```

```python
#calculating missing rate for each column and sorting
missing_rate = (df.isnull().sum() * 100) / df.shape[0]
missing_rate.sort_values(ascending=False)
```

```python
#dropping columns with a high null value rate
columns_to_drop = ['OWN_CAR_AGE', 'OCCUPATION_TYPE']
df.drop(columns=columns_to_drop, inplace=True)
```

```python
#calculating missing rate for each column and sorting
missing_rate = (df.isnull().sum() * 100) / df.shape[0]
missing_rate.sort_values(ascending=False)
```

```python
#checking for number of unique values in each of categorical columns
df.select_dtypes(include='object').nunique()
```

```python
#dataframe correlation matrix
correlation_matrix = df.corr()
correlation_df = pd.DataFrame(correlation_matrix)
correlation_df
```

```python
#dropping columns which are correlated less than 0.01 with TARGET column
correlation_with_target = df.drop('SK_ID_CURR', axis=1).corr()['TARGET'].abs()
columns_to_drop = correlation_with_target[correlation_with_target < 0.01].index

df.drop(columns=columns_to_drop, inplace=True)
```

```python
#filling rest of the null values
df['AMT_REQ_CREDIT_BUREAU_YEAR'].fillna(0, inplace=True)
df['AMT_REQ_CREDIT_BUREAU_MON'].fillna(0, inplace=True)
df['DEF_60_CNT_SOCIAL_CIRCLE'].fillna(0, inplace=True)
df['DEF_30_CNT_SOCIAL_CIRCLE'].fillna(0, inplace=True)
df['NAME_TYPE_SUITE'].fillna('Unknown', inplace=True)
df['AMT_GOODS_PRICE'].fillna(df['AMT_GOODS_PRICE'].mean(), inplace=True)
df['AMT_ANNUITY'].fillna(df['AMT_ANNUITY'].mean(), inplace=True)
df['EXT_SOURCE_1'].fillna(df['EXT_SOURCE_1'].mean(), inplace=True)
df['EXT_SOURCE_2'].fillna(df['EXT_SOURCE_2'].mean(), inplace=True)
df['EXT_SOURCE_3'].fillna(df['EXT_SOURCE_3'].mean(), inplace=True)
```

```python
#dropping 1 row with a missing value
df = df.dropna(subset=['DAYS_LAST_PHONE_CHANGE'], axis=0)
```
