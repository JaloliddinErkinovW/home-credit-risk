### bureau

```python
#loading other tables in order to merge all of them
bureau = pd.read_csv('data/cleaned/bureau_cleaned.csv')
bureau_balance = pd.read_csv('data/cleaned/bureau_balance_cleaned.csv')
```

```python
#merging tables
new_bureau = pd.merge(bureau, bureau_balance, on = 'SK_ID_BUREAU', how = 'inner')

del bureau
del bureau_balance

new_bureau.head()
```

```python
#aggregation operations
agg_data = {
    'SK_ID_BUREAU': 'count',
    'CREDIT_ACTIVE': lambda x: (x == 'Active').sum(),
    'CREDIT_CURRENCY': lambda x: x.mode().iloc[0],
    'DAYS_CREDIT' : 'mean',
    'CREDIT_DAY_OVERDUE': 'mean',
    'DAYS_CREDIT_ENDDATE': 'mean',
    'DAYS_ENDDATE_FACT': 'mean',
    'AMT_CREDIT_MAX_OVERDUE': 'max',
    'CNT_CREDIT_PROLONG': 'sum',
    'AMT_CREDIT_SUM': 'sum',
    'AMT_CREDIT_SUM_DEBT': 'sum',
    'AMT_CREDIT_SUM_LIMIT': 'sum',
    'AMT_CREDIT_SUM_OVERDUE': 'sum',
    'CREDIT_TYPE': lambda x: x.mode().iloc[0],
    'DAYS_CREDIT_UPDATE': 'mean',
    'AMT_ANNUITY': 'sum',
    'MONTHS_BALANCE' : 'mean'
}
```

```python
#applying aggregation
new_bureau = new_bureau.groupby('SK_ID_CURR').agg(agg_data)
```

```python
new_bureau.shape
```

```python
#merging tables
new_train = pd.merge(train, new_bureau, on = 'SK_ID_CURR', suffixes=('', '_bureau'), how = 'left')

del train
del new_bureau

new_train.head()
```

```python
new_train.shape
```

```python
new_train.isnull().sum()
```

```python
columns_to_drop = ['SK_ID_BUREAU']
new_train.drop(columns=columns_to_drop, inplace=True)
```

### installments_payments

```python
#loading dataset into dataframe
installments_payments = pd.read_csv('data/cleaned/installments_payments_cleaned.csv')
```

```python
#creating a new column DAYS_NP (test)
installments_payments['DAYS_NP'] = installments_payments['DAYS_ENTRY_PAYMENT'] - installments_payments['DAYS_INSTALMENT']
```

```python
installments_payments.head()
```

```python
#creating a new column NPL_SEEN (test)
installments_payments['NPL_SEEN'] = installments_payments['DAYS_NP'].apply(lambda x: 1 if x < -90 else 0)
```

```python
installments_payments.head()
```

```python
#aggregation operations
agg_data = {
    'SK_ID_PREV': 'count',
    'DAYS_INSTALMENT' : 'max',
    'DAYS_ENTRY_PAYMENT': 'max',
    'DAYS_NP': 'min',
    'AMT_INSTALMENT': 'mean',
    'AMT_PAYMENT': 'mean',
    'NPL_SEEN': 'max'
}
```

```python
#applying aggregation
installments_payments = installments_payments.groupby('SK_ID_CURR').agg(agg_data)
```

```python
#merging tables
final_table = pd.merge(new_train, installments_payments, on = 'SK_ID_CURR', suffixes=('', '_installment'), how = 'left')

del new_train
del installments_payments

final_table.head()
```

```python
final_table.shape
```

```python
final_table.isnull().sum()
```

```python
#filling null values and dropping ID columns which is not needed anymore
final_table['DAYS_INSTALMENT'].fillna(0, inplace=True)
final_table['DAYS_ENTRY_PAYMENT'].fillna(0, inplace=True)
final_table['DAYS_NP'].fillna(0, inplace=True)
final_table['AMT_INSTALMENT'].fillna(0, inplace=True)
final_table['AMT_PAYMENT'].fillna(0, inplace=True)
final_table['NPL_SEEN'].fillna(0, inplace=True)

columns_to_drop = ['SK_ID_PREV']
final_table.drop(columns=columns_to_drop, inplace=True)
```

```python
final_table.isnull().sum()
```

### previous_application

```python
#loading dataset into dataframe
pa = pd.read_csv('data/cleaned/previous_application_cleaned.csv')

pa.head()
```

```python
#aggregation operations
agg_data = {
    'SK_ID_PREV': 'count',
    'AMT_ANNUITY' : 'mean',
    'AMT_APPLICATION': 'mean',
    'AMT_CREDIT': 'mean',
    'AMT_DOWN_PAYMENT': 'mean',
    'AMT_GOODS_PRICE': 'mean',
    'DAYS_TERMINATION': 'mean',
    'DAYS_LAST_DUE': 'mean',
    'DAYS_LAST_DUE_1ST_VERSION': 'mean',
    'DAYS_FIRST_DUE': 'mean',
    'DAYS_FIRST_DRAWING': 'mean',
    'CNT_PAYMENT': 'mean'
}
```

```python
#applying aggregation
pa = pa.groupby('SK_ID_CURR').agg(agg_data)
```

```python
pa.shape
```

```python
#merging tables
df = pd.merge(final_table, pa, on = 'SK_ID_CURR', suffixes=('', '_prev'), how = 'left')

del final_table
del pa
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
#filling all numerical columns' missing values with 0
numerical_columns = df.select_dtypes(exclude=['object']).columns
df[numerical_columns] = df[numerical_columns].fillna(0)
```

```python
df.isnull().sum()
```

```python
#dropping columns which are not necessary
columns_to_drop = ['CREDIT_CURRENCY', 'CREDIT_TYPE']
df = df.drop(columns=columns_to_drop)
```
