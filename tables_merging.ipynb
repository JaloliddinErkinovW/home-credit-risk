```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
```

```python
#loading the table
train = pd.read_csv('data/application_train.csv')

train.head()
```

```python
train.shape
```

```python
train.describe()
```

```python
#table info
train_dtypes_nulls = pd.DataFrame({
    'missing_bureau': train.isnull().sum(),
    'data_types_bureau': train.dtypes
})

train_dtypes_nulls
```

```python
#filling na values
for column in train.columns:
    if 'AVG' in column:
        train[column].fillna(train[column].mean(), inplace=True)
    elif 'MEDI' in column:
        train[column].fillna(train[column].median(), inplace=True)
    elif 'MODE' in column:
        train[column].fillna(train[column].mode()[0], inplace=True)
```

```python
train.isnull().sum()
```

```python
#dropping columns with more than 100 null rows
columns_to_drop = train.columns[(train.isnull().sum() > 100)]

train = train.drop(columns=columns_to_drop)
```

```python
train.isnull().sum()
```

### Merging bureau

```python
#loading other tables in order to merge all of them
bureau = pd.read_csv('data/cleaned/bureau_cleaned.csv')
bureau_balance = pd.read_csv('data/cleaned/bureau_balance_cleaned.csv')
```

```python
#merging tables
new_bureau = pd.merge(bureau, bureau_balance, on = 'SK_ID_BUREAU', how = 'inner')

del bureau
del bureau_balance

new_bureau.head()
```

```python
#aggregation operations
agg_data = {
    'SK_ID_BUREAU': 'count',
    'CREDIT_ACTIVE': lambda x: (x == 'Active').sum(),
    'CREDIT_CURRENCY': lambda x: x.mode().iloc[0],
    'DAYS_CREDIT' : 'mean',
    'CREDIT_DAY_OVERDUE': 'mean',
    'DAYS_CREDIT_ENDDATE': 'mean',
    'DAYS_ENDDATE_FACT': 'mean',
    'AMT_CREDIT_MAX_OVERDUE': 'max',
    'CNT_CREDIT_PROLONG': 'sum',
    'AMT_CREDIT_SUM': 'sum',
    'AMT_CREDIT_SUM_DEBT': 'sum',
    'AMT_CREDIT_SUM_LIMIT': 'sum',
    'AMT_CREDIT_SUM_OVERDUE': 'sum',
    'CREDIT_TYPE': lambda x: x.mode().iloc[0],
    'DAYS_CREDIT_UPDATE': 'mean',
    'AMT_ANNUITY': 'sum',
    'MONTHS_BALANCE' : 'mean'
}
```

```python
#applying aggregation
new_bureau = new_bureau.groupby('SK_ID_CURR').agg(agg_data)
```

```python
new_bureau.shape
```

```python
#merging tables
new_train = pd.merge(train, new_bureau, on = 'SK_ID_CURR', how = 'inner')

del train
del new_bureau

new_train.head()
```

```python
new_train.shape
```

```python
new_train.isnull().sum()
```

```python
#saving table into csv
new_train.to_csv('data/cleaned/application_train_cleaned.csv', index=False)
```
