### application_test merging

```python
df = pd.read_csv('data/cleaned/application_test_cleaned.csv')

df.head()
```

```python
columns_to_drop = ['Unnamed: 0']
df.drop(columns=columns_to_drop, inplace=True)
```

```python
bureau = pd.read_csv('data/cleaned/bureau_cleaned.csv')
bureau_balance = pd.read_csv('data/cleaned/bureau_balance_cleaned.csv')
```

```python
new_bureau = pd.merge(bureau, bureau_balance, on = 'SK_ID_BUREAU', how = 'inner')

del bureau
del bureau_balance
```

```python
agg_data = {
    'SK_ID_BUREAU': 'count',
    'CREDIT_ACTIVE': lambda x: (x == 'Active').sum(),
    'CREDIT_CURRENCY': lambda x: x.mode().iloc[0],
    'DAYS_CREDIT' : 'mean',
    'CREDIT_DAY_OVERDUE': 'mean',
    'DAYS_CREDIT_ENDDATE': 'mean',
    'DAYS_ENDDATE_FACT': 'mean',
    'AMT_CREDIT_MAX_OVERDUE': 'max',
    'CNT_CREDIT_PROLONG': 'sum',
    'AMT_CREDIT_SUM': 'sum',
    'AMT_CREDIT_SUM_DEBT': 'sum',
    'AMT_CREDIT_SUM_LIMIT': 'sum',
    'AMT_CREDIT_SUM_OVERDUE': 'sum',
    'CREDIT_TYPE': lambda x: x.mode().iloc[0],
    'DAYS_CREDIT_UPDATE': 'mean',
    'AMT_ANNUITY': 'sum',
    'MONTHS_BALANCE' : 'mean'
}

new_bureau = new_bureau.groupby('SK_ID_CURR').agg(agg_data)
```

```python
new_test = pd.merge(df, new_bureau, on = 'SK_ID_CURR', suffixes=('', '_bureau'), how = 'left')

del df
del new_bureau
```

```python
columns_to_drop = ['SK_ID_BUREAU']
new_test.drop(columns=columns_to_drop, inplace=True)
```

```python
installments_payments = pd.read_csv('data/cleaned/installments_payments_cleaned.csv')
```

```python
installments_payments['DAYS_NP'] = installments_payments['DAYS_ENTRY_PAYMENT'] - installments_payments['DAYS_INSTALMENT']
installments_payments['NPL_SEEN'] = installments_payments['DAYS_NP'].apply(lambda x: 1 if x < -90 else 0)
```

```python
agg_data = {
    'SK_ID_PREV': 'count',
    'DAYS_INSTALMENT' : 'max',
    'DAYS_ENTRY_PAYMENT': 'max',
    'DAYS_NP': 'min',
    'AMT_INSTALMENT': 'mean',
    'AMT_PAYMENT': 'mean',
    'NPL_SEEN': 'max'
}

installments_payments = installments_payments.groupby('SK_ID_CURR').agg(agg_data)
```

```python
newest_test = pd.merge(new_test, installments_payments, on = 'SK_ID_CURR', suffixes=('', '_installment'), how = 'left')

del new_test
del installments_payments
```

```python
newest_test['DAYS_INSTALMENT'].fillna(0, inplace=True)
newest_test['DAYS_ENTRY_PAYMENT'].fillna(0, inplace=True)
newest_test['DAYS_NP'].fillna(0, inplace=True)
newest_test['AMT_INSTALMENT'].fillna(0, inplace=True)
newest_test['AMT_PAYMENT'].fillna(0, inplace=True)
newest_test['NPL_SEEN'].fillna(0, inplace=True)

columns_to_drop = ['SK_ID_PREV']
newest_test.drop(columns=columns_to_drop, inplace=True)
```

```python
pa = pd.read_csv('data/cleaned/previous_application_cleaned.csv')
```

```python
agg_data = {
    'SK_ID_PREV': 'count',
    'AMT_ANNUITY' : 'mean',
    'AMT_APPLICATION': 'mean',
    'AMT_CREDIT': 'mean',
    'AMT_DOWN_PAYMENT': 'mean',
    'AMT_GOODS_PRICE': 'mean',
    'DAYS_TERMINATION': 'mean',
    'DAYS_LAST_DUE': 'mean',
    'DAYS_LAST_DUE_1ST_VERSION': 'mean',
    'DAYS_FIRST_DUE': 'mean',
    'DAYS_FIRST_DRAWING': 'mean',
    'CNT_PAYMENT': 'mean'
}

pa = pa.groupby('SK_ID_CURR').agg(agg_data)
```

```python
final_test = pd.merge(newest_test, pa, on = 'SK_ID_CURR', suffixes=('', '_prev'), how = 'left')

del newest_test
del pa
```

```python
numerical_columns = final_test.select_dtypes(exclude=['object']).columns
final_test[numerical_columns] = final_test[numerical_columns].fillna(0)
```

```python
columns_to_drop = ['CREDIT_CURRENCY', 'CREDIT_TYPE']
final_test = final_test.drop(columns=columns_to_drop)
```

```python
final_test.to_csv('data/cleaned/test_merged.csv', index=False)
```

### Predicting

```python
df = pd.read_csv('data/cleaned/test_filtered.csv')

df.head()
```

```python
selected_columns = ['CODE_GENDER', 'DAYS_BIRTH', 'DAYS_EMPLOYED', 'AMT_CREDIT', 'AMT_PAYMENT', 'AMT_INSTALMENT', 'AMT_GOODS_PRICE']
df_selected = df[selected_columns]
df_selected.shape
```

```python
categorical_columns = df_selected.select_dtypes(include=['object']).columns

df_encoded = pd.get_dummies(df_selected, columns=categorical_columns)

df_encoded.head()
```

```python
df_encoded.shape
```

```python
# df_encoded['CODE_GENDER_XNA'] = 0
# df_encoded['NAME_FAMILY_STATUS_Unknown'] = 0
# df_encoded['NAME_INCOME_TYPE_Maternity'] = 0
df_encoded['CODE_GENDER_XNA'] = 0
```

```python
df_encoded.shape
```

```python
X_test = df_encoded

model_name = "Gradient Boosting"
model_path = f"models/{model_name}_selected.joblib"
model = joblib.load(model_path)

feature_names = X_test.columns

X_test = X_test.reindex(columns=feature_names, fill_value=0)

predictions = model.predict(X_test)

df['TARGET_PR'] = predictions

df.to_csv('data/cleaned/predicted_test.csv', index=False)
```
